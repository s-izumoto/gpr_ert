```python
import numpy as np
import pandas as pd

npz_path = "seq_logs_bundle.npz"  # ←必要に応じて変更

def small_preview(a, max_items=6):
    """配列の冒頭だけ文字列でプレビュー"""
    if a.ndim == 0:
        # スカラー
        try:
            return repr(a.item())
        except Exception:
            return str(a)
    if a.ndim == 1:
        return np.array2string(a[:max_items], precision=4, threshold=max_items)
    if a.ndim == 2:
        r = min(3, a.shape[0])
        c = min(6, a.shape[1])
        return np.array2string(a[:r, :c], precision=4, threshold=r*c)
    # 3次元以上は先頭スライスを表示
    head = a[(0,) * (a.ndim - 2)]
    r = min(3, head.shape[0])
    c = min(6, head.shape[1])
    return f"slice[0,...]:\n" + np.array2string(head[:r, :c], precision=4, threshold=r*c)

with np.load(npz_path, allow_pickle=False) as z:
    rows = []
    for k in z.files:
        a = z[k]
        size_mb = round(a.nbytes / (1024**2), 3)
        rows.append((k, a.shape, str(a.dtype), size_mb, small_preview(a)))

df = pd.DataFrame(rows, columns=["key", "shape", "dtype", "size_MB", "preview"]).sort_values("key")
df
```

```python
import numpy as np
import pandas as pd

# ---------- 設定 ----------
npz_path = "medoids_index.npz"   # ここを必要に応じて変更
target_key = "cluster"         # ← ここに全値を見たいキー名（例: 'src_index', 'cluster', 'order', 'png_relpath'）
save_csv = True                  # TrueならCSV保存（ファイル名は {target_key}.csv）

# ---------- 便利関数 ----------
def small_preview(a, max_items=6):
    """配列の冒頭だけ文字列でプレビュー"""
    if a.ndim == 0:
        try:
            return repr(a.item())
        except Exception:
            return str(a)
    if a.ndim == 1:
        return np.array2string(a[:max_items], precision=4, threshold=max_items)
    if a.ndim == 2:
        r = min(3, a.shape[0])
        c = min(6, a.shape[1])
        return np.array2string(a[:r, :c], precision=4, threshold=r*c)
    # 3次元以上は先頭スライスを表示
    head = a[(0,) * (a.ndim - 2)]
    r = min(3, head.shape[0])
    c = min(6, head.shape[1])
    return f"slice[0,...]:\n" + np.array2string(head[:r, :c], precision=4, threshold=r*c)

# ---------- 読み込み＆一覧 ----------
with np.load(npz_path, allow_pickle=True) as z:  # png_relpath が object/文字列でも確実に読むため allow_pickle=True
    rows = []
    for k in z.files:
        a = z[k]
        size_mb = round(a.nbytes / (1024**2), 3) if hasattr(a, "nbytes") else 0.0
        rows.append((k, a.shape, str(a.dtype), size_mb, small_preview(a)))

    df = pd.DataFrame(rows, columns=["key", "shape", "dtype", "size_MB", "preview"]).sort_values("key")
    print("=== Keys overview ===")
    print(df.to_string(index=False))

    # ---------- 指定キーの全値を表示 ----------
    if target_key:
        if target_key not in z.files:
            raise KeyError(f"Key '{target_key}' not found. Available keys: {list(z.files)}")
        arr = z[target_key]
        print(f"\n=== FULL values for key='{target_key}' ===")
        print(f"shape={arr.shape} dtype={arr.dtype}")

        if arr.ndim == 0:
            # スカラー
            print(arr.item())
            if save_csv:
                pd.Series([arr.item()], name=target_key).to_csv(f"{target_key}.csv", index=False, encoding="utf-8")
                print(f"[saved] {target_key}.csv")
        elif arr.ndim == 1:
            # 1次元は全部表示
            # 長い場合は print が溢れるので、必要なら head/tail に切り替えてください
            print(arr)
            if save_csv:
                pd.Series(arr, name=target_key).to_csv(f"{target_key}.csv", index=False, encoding="utf-8")
                print(f"[saved] {target_key}.csv")
        else:
            # 2次元以上は DataFrame 化（最後の次元で展開）
            flat = arr.reshape(arr.shape[0], -1)
            df_all = pd.DataFrame(flat)
            # 表示は head のみ（フルはCSVで）
            print(df_all.head(10).to_string(index=False))
            if save_csv:
                df_all.to_csv(f"{target_key}.csv", index=False, encoding="utf-8")
                print(f"[saved] {target_key}.csv")
```
