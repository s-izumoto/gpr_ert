```python
import numpy as np, hashlib, json

def row_hash(vec: np.ndarray) -> int:
    v = np.ascontiguousarray(vec).view(np.uint8)
    return int(hashlib.sha1(v).hexdigest()[:8], 16)  # 32bit相当

def field_hashes(npz_path: str, key="Z"):
    with np.load(npz_path, allow_pickle=True) as d:
        meta = {}
        if "meta" in d.files:
            try: meta = json.loads(d["meta"][0])
            except Exception: pass
        R = int(meta.get("rows_per_field", 0)) or None
        Z = d[key] if key in d.files else None
        if Z is None or R is None:
            raise RuntimeError(f"need {key} and rows_per_field in {npz_path}")
        N_rows = Z.shape[0]; nF = N_rows // R
        hs = np.empty(nF, dtype=np.int64)
        for f in range(nF):
            hs[f] = row_hash(Z[f*R])    # フィールド先頭行で代表
        return hs

A = r"..\data\interim\ert_wenner_subset\ert_surrogate_wenner_reduced.npz"
B = r"..\data\interim\surrogate_ds\ert_surrogate_reduced.npz"

hA = field_hashes(A); hB = field_hashes(B)
print("n_fields A,B:", len(hA), len(hB))
print("same multiset? ->", np.array_equal(np.sort(hA), np.sort(hB)))

# A の順番に並ぶよう、B 側の並び替え perm_B を作る
posB = {int(h): i for i, h in enumerate(hB)}
perm_B = np.array([posB.get(int(h), -1) for h in hA], dtype=int)
missing = int((perm_B < 0).sum())
print("unmatched (should be 0):", missing)
```

```python
# ==== Quick sanity check for ERTOracleDataset ====
import torch
from torch.utils.data import DataLoader
from toy_evidential_set_transformer import HP, ERTOracleDataset, collate_fn

# --- ここは run_train と同じ設定に揃える ---
ERT_NPZ    = "../data/interim/surrogate_ds/ert_surrogate_reduced.npz"
ORACLE_NPZ = "../data/processed/oracle_pairs_reduced/oracle_pairs_reduced_diverse.npz"
WARMUP_NPZ = "../data/interim/ert_wenner_subset/ert_surrogate_wenner_reduced.npz"

# --- Dataset を最少フィールドで作成（速い）---
ds = ERTOracleDataset(
    ert_npz_path=ERT_NPZ,
    oracle_npz_path=ORACLE_NPZ,
    hp=HP, pick_scale_index=0,
    max_fields=2, field_offset=0,           # ← 手早く確認したいので2件だけ
    warmup_npz_path=WARMUP_NPZ,
    warmup_pick="first35",
    active_select="random",
    active_seed=20251021
)

print(f"len(ds) = {len(ds)}  | ROI n={ds.n_roi}  | grid=({HP.grid_h},{HP.grid_w})")
print("-"*60)

def check_one(idx=0):
    sets, y = ds[idx]   # sets:[T,L,5], y:[n_roi]
    T, L, D = sets.shape
    print(f"[sample #{idx}] sets.shape={sets.shape}, y.shape={y.shape}")

    # --- 形チェック ---
    assert D == 5, "last dim must be 5 = [4D design + y]"
    assert T == HP.warmup_fixed + HP.active_len, f"T must be warmup_fixed+active_len ({HP.warmup_fixed}+{HP.active_len})"
    assert y.shape[0] == ds.n_roi, "y length must match ROI size"

    # --- 値レンジざっくり（NaN/infの検出も）---
    assert torch.isfinite(sets).all(), "sets has NaN/Inf"
    assert torch.isfinite(y).all(), "y has NaN/Inf"
    print(f"  sets min/max: {sets.min().item():.4g} / {sets.max().item():.4g}")
    print(f"  y   uniq: {torch.unique(y).tolist()} (binary expected: [0.0, 1.0] or subset)")

    # --- Warm-up/Activeの長さ表示 ---
    print(f"  warmup_fixed={HP.warmup_fixed}, active_len={HP.active_len}")

    # --- L複製の確認（時刻tで L 個は完全に同一か？）---
    # 代表として t=0, t=warmup_fixed-1, t=warmup_fixed, t=T-1 をチェック
    for t in {0, max(0,HP.warmup_fixed-1), HP.warmup_fixed, T-1}:
        block = sets[t]              # [L,5]
        diffs = (block - block[0]).abs().max().item()
        print(f"  t={t:>3d}  replicate L check: max(abs diff)={diffs:.3g}")
        # 厳格に同一を期待するなら:
        # assert diffs < 1e-6, f"L copies at t={t} are not identical"

    # --- 先頭/中間/末尾の v5 を少しだけ覗く（デバッグ用のサンプル表示）---
    for t in [0, HP.warmup_fixed-1, HP.warmup_fixed, T-1]:
        v5 = sets[t, 0]  # その時刻の代表（L複製の先頭）
        dnorm = v5[:4].tolist()
        y_meas = float(v5[4].item())
        print(f"  t={t:>3d} v5 = [Dnorm={list(map(lambda x: round(x,4), dnorm))}, log10(rhoa)={y_meas:.4f}]")

check_one(1)
print("-"*60)

# --- DataLoader 経由でバッチも形だけ確認 ---
dl = DataLoader(ds, batch_size=2, shuffle=False, drop_last=False, collate_fn=collate_fn)
sets_b, y_b = next(iter(dl))  # sets_b:[B,T,L,5], y_b:[B,n_roi]
print(f"[batch] sets.shape={sets_b.shape}, y.shape={y_b.shape}")
assert sets_b.shape[1] == HP.warmup_fixed + HP.active_len
assert sets_b.shape[3] == 5
assert y_b.shape[1] == ds.n_roi
print("OK: dataset outputs look consistent.")
```

```python
# --- 可視化用: 1つの y（ROIバイナリ）を画像として出力 ---
import numpy as np
import torch
import matplotlib.pyplot as plt
from pathlib import Path

# 1) Dataset を用意（パスはあなたの環境に合わせて）
ERT_NPZ    = "../data/interim/surrogate_ds/ert_surrogate_reduced.npz"
ORACLE_NPZ = "../data/processed/oracle_pairs_reduced/oracle_pairs_reduced_diverse.npz"
WARMUP_NPZ = "../data/interim/ert_wenner_subset/ert_surrogate_wenner_reduced.npz"

ds = ERTOracleDataset(
    ert_npz_path=ERT_NPZ,
    oracle_npz_path=ORACLE_NPZ,
    hp=HP, pick_scale_index=0,
    max_fields=1, field_offset=0,    # ← 1件だけでOK
    warmup_npz_path=WARMUP_NPZ,
    warmup_pick="first35",
    active_select="random",
    active_seed=20251021
)

# 2) 1サンプル取り出し
sets, y = ds[0]                    # sets:[T,L,5], y:[n_roi]
y = y.detach().cpu().numpy()       # [n_roi]

# 3) ROIインデックスを full grid に戻す
gh, gw = HP.grid_h, HP.grid_w
canvas = np.full((gh * gw,), np.nan, dtype=np.float32)  # ROI外はNaNにして目立たせる
canvas[ds.roi_idx] = y.astype(np.float32)               # ROI内に値を再配置
img = canvas.reshape(gh, gw)                            # [gh, gw]

# 4) 表示（NaN は空白っぽく見える）
plt.figure()
plt.imshow(img, origin="upper")
plt.title(f"Oracle y (field #0), grid=({gh},{gw}), ROI count={ds.n_roi}")
plt.colorbar()
plt.tight_layout()

# 5) 保存（任意）
out_path = Path("./y_example_field0.png")
plt.savefig(out_path, dpi=150)
print(f"Saved: {out_path.resolve()}")
```
